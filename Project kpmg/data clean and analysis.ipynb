```python
import pandas as pd
from datetime import datetime

# Load the Excel file with multiple sheets
file_path = r'C:\Users\sakif\KPMG_VI_Analysis.xlsx'
xls = pd.ExcelFile(file_path)

# Load the sheets into DataFrames
transactions = pd.read_excel(xls, 'Transactions')
customer_demographic = pd.read_excel(xls, 'CustomerDemographic')
customer_address = pd.read_excel(xls, 'CustomerAddress')
new_customer_list = pd.read_excel(xls, 'NewCustomerList')

# Define current date for recency calculation
current_date = datetime.now()

# Step 1: Calculate Profit (Monetary) for each transaction
transactions['profit'] = transactions['list_price'] - transactions['standard_cost']

# Step 2: Calculate Recency for each customer
# Convert transaction_date to datetime
transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'])

# Find the most recent transaction date for each customer
recency_df = transactions.groupby('customer_id').agg({
    'transaction_date': lambda x: (current_date - x.max()).days
}).reset_index()

recency_df.rename(columns={'transaction_date': 'recency'}, inplace=True)

# Step 3: Calculate Frequency for each customer (total number of transactions)
frequency_df = transactions.groupby('customer_id').agg({
    'transaction_id': 'count'
}).reset_index()

frequency_df.rename(columns={'transaction_id': 'frequency'}, inplace=True)

# Step 4: Calculate Monetary (sum of profits for each customer)
monetary_df = transactions.groupby('customer_id').agg({
    'profit': 'sum'
}).reset_index()

monetary_df.rename(columns={'profit': 'monetary'}, inplace=True)

# Step 5: Merge Recency, Frequency, and Monetary into a single DataFrame
rfm_df = recency_df.merge(frequency_df, on='customer_id').merge(monetary_df, on='customer_id')

# Step 6: Optional - Add customer demographic data for further analysis
# Merge with customer demographic on customer_id
rfm_df = rfm_df.merge(customer_demographic[['customer_id', 'wealth_segment']], on='customer_id', how='left')

# Step 7: Rank customers based on Recency, Frequency, and Monetary
# You can use quantiles or a scoring system (1-5)

# Example: Using quantiles for ranking
rfm_df['recency_score'] = pd.qcut(rfm_df['recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_df['frequency_score'] = pd.qcut(rfm_df['frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm_df['monetary_score'] = pd.qcut(rfm_df['monetary'], 5, labels=[1, 2, 3, 4, 5])

# Step 8: Calculate the overall RFM score
rfm_df['RFM_score'] = rfm_df['recency_score'].astype(int) + rfm_df['frequency_score'].astype(int) + rfm_df['monetary_score'].astype(int)

# Step 9: Sort the customers based on the RFM score
rfm_df.sort_values(by='RFM_score', ascending=False, inplace=True)

# Output the RFM result
print(rfm_df.head())


```

```python
import pandas as pd

# Load the RFM analysis result file
file_path = r'C:\Users\sakif\rfm_analysis_result.xlsx'
rfm_df = pd.read_excel(file_path)

# Define a function to categorize customers into four groups based on RFM score
def rfm_segment(score):
    if score >= 11:
        return 'High-Value Customers (11-15)'
    elif score >= 6:
        return 'Potential Loyalists (6-10)'
    elif score >= 4:
        return 'At-Risk Customers (4-6)'
    else:
        return 'Low-Value Customers (1-3)'

# Apply the function to create a new column 'RFM_Segment'
rfm_df['RFM_Segment'] = rfm_df['RFM_score'].apply(rfm_segment)

# View the updated DataFrame with the new RFM segments
print(rfm_df[['customer_id', 'RFM_score', 'RFM_Segment']].head())
```
